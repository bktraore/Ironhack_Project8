```python
df['Dates'] = pd.to_datetime(df['Date/Time']).dt.date
df['Time'] = pd.to_datetime(df['Date/Time']).dt.hour
```

```python
firstindex  = df.pop("Dates")
df.insert(0, 'Dates', firstindex)
secondindex = df.pop("Time")
df.insert(1, 'Time', secondindex)

df = df.drop(['Date/Time'], axis=1)
```

```python
df['month'] = pd.to_datetime(df['Dates']).dt.month
df['month_name'] = pd.to_datetime(df['Dates']).dt.month_name()
df['Day_of_month'] = pd.to_datetime(df['Dates']).dt.day
df['Day_name'] = pd.to_datetime(df['Dates']).dt.day_name()
df['Day_of_weeks'] = pd.to_datetime(df['Dates']).dt.day_of_week
```

```python
df
```

```python
df.dtypes
```

```python
bins = [0, 6, 11, 15, 18, 23]
labels = ['night', 'morning', 'midday','afternoon','evening']
df['period'] = pd.cut(x = df['Time'], bins = bins, labels = labels, include_lowest = True)
df
```

```python
bins = [0,4,6]
labels = ['workweek','weekend']
df['weekend?'] = pd.cut(x = df['Day_of_weeks'], bins = bins, labels = labels, include_lowest = True)
df
```
